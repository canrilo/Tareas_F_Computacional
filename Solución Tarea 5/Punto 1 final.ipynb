```python
"""
Programa que calcula la transformada de fourier de un encefalograma
Adicionalmente calcula los espectros de potencias, y reconstruye la señal a partir de los 10 espectros mas altos

Autores: Camilo Rivera
         Nancy Ruiz
Fecha de creación: Octubre 9 de 2013
"""


%pylab inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
from scipy.fftpack import fft, fftfreq
```

```python
#metodo que importa el archivo
#loadtxt es una funcion de numpy que lee todo como int
datos = loadtxt('sampled+ma0844az_1-1+_data.txt')
```

```python
n=len(datos)
```

```python
#numero de columnas o encefalogramas
N=24
```

```python
#plot de todos los encefalogramas

for i in range(N):
    vector=datos[:,i]
    plt.plot(vector)
```

```python
"""
Metodo que calcula fft de cada señal
Input: 24 vectores
Output: fft de cada vector
"""
#El espacio entre cada senal es 1 minuto
dt=1

#Matriz para las frecuencias de fourier
freq=np.zeros(shape=(400,24),dtype=float) 

#matriz de fft para cada señal (tipo complex)
matrizTransformadas=np.zeros(shape=(400,24),dtype=numpy.complex)

for i in range(N):
    senal=datos[:,i]
    fft_x = fft(senal) / n # FFT Normalized
    matrizTransformadas[:,i]=fft_x
    fft_z=fftfreq(n,dt)
    freq[:,i]=fft_z
    
    
for i in range(N):
    matrizTransformadas[:,i]=np.fft.fftshift(matrizTransformadas[:,i])
    freq[:,i]=np.fft.fftshift(freq[:,i])
```

```python
"""
Metodo que calcula la norma al cuadrado de cada transformada de fourier
y hace una grafica del espectro de potencias
"""
#matriz de valor absoluto
pot=np.zeros((n,N))
for i in range(n):
    for j in range(N):
        pot[i,j]=np.abs(matrizTransformadas[i,j])
        


#matriz de cuadrados del valor absoluto
cuadrado=np.zeros((n,N))
for i in range(n):
    for j in range(N):
        cuadrado[i,j]=pot[i,j]**2
    
print len(cuadrado)

#plot de las potencias
for i in range (N):
    potencias=cuadrado[:,i]
    plt.semilogy(freq[:,0],potencias)
    title('Grafica de Potencia en funcion de la frecuencia')
```

```python
"""
Metodo que detecta los 10 picos más altos y los guarda en un array
input: espectros de potencias:cuadrado
output: 10 valores más altos:mas_altos
"""


def mas_altos(entrada,compleja):
    #Se crean los vectores donde se almacenarán los 10 mayores y sus índices
    mayores=np.zeros((10,1))
    #Para el vector de índices se inicializa con -1 para evitar problemas con la posición cero
    indices=np.array([-1 for w in range(10)])
    #Se procede a iterar sobre cada punto de la señal de entrada (Excepto el primero y el último que ya se miraron)
    for i in range(27):
        #Se compara con cada uno de los valores en el vector de mayores
        for j in range(10):
            if(entrada[i]>mayores[j]):
                #En caso de que sea mayor que el valor actual, se procede a desplazar el resto de valores
                #Se realiza el corrimiento en orden inverso para no sobreescribir valores
                for k in range(9,j,-1):
                    mayores[k]=mayores[k-1]
                    indices[k]=indices[k-1]
                #Se escribe el nuevo valor en el vector y el índice correspondiente y se detiene el loop
                mayores[j]=entrada[i]
                indices[j]=i
                break
    
    #Se realiza un nuevo recorrido para dejar sólo los mayores
    for i in range(27):
        m=0 #Variable auxiliar que permitirá saber si el índice actual se encuentra en la lista
        for j in range(10):
            #Si el índice actual se encuentra en la lista se cambia la variable y se acaba el loop
            if(i==indices[j]):
                m=1
                break
        #En caso de no encontrarse en toda la lista se reemplaza el valor actual por cero
        if(m==0):
            compleja[i]=0
    
    #Se retorna el vector corregido
    return compleja


#vectores de 10 valores mas altos se va a llamar cuadrados_ceros
cuadrados_ceros=np.zeros((400,24),dtype=complex)
for i in range(N):
    cuadrados_ceros[:,i]= mas_altos(cuadrado[:,i],matrizTransformadas[:,i])

```
